1. Створити DataFrame для розрахунку заробітної плати за поточний місяць за даними: ПІБ, дата народження, посадовий оклад, стаж, кількість відпрацьованих днів. Таблицю розробити для 10 співробітників (рядків).
2. Для формування DataFrame розрахунку заробітньої плати необхідно:
2.1. В створений DataFrame додати колонки:
2.1.1. розрахунку нарахованої з/п за формулою: нарахована з/п = посадовий оклад / кількість робочих днів місяця * кількість відпрацьованих днів; Кількість робочих днів поточного місяця отримати за посиланням: https://buhgalter911.com/uk/proizvodstvennyj-calendar/.
 2.1.2. доплата за стаж: за кожний рік стажу 0,1% від посадового окладу;
2.1.3. Всього нараховано=нарахована з/п + доплата за стаж;
2.1.4. Податок = всього нараховано -19,5%;
2.1.5. До видачі = всього нараховано-податок.
 2.2. Зберегти розрахований DataFrame.
2.3. Використовуючи отримані дані провести статистичні розрахунки: визначити середню з/п співробітників які пропрацювали повний місяць, скільки співробітників мають стаж більше 10 років. Результат вивести на екран.
 2.4. Візуалізувати розраховану з/п співробітників. За результатами виконаної роботи оформити звіт в форматі pdf


```python

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# 1. Створення DataFrame
data = {
    'ПІБ': ['Іванов Іван Петрович', 'Петренко Петро Георгійович', 'Калина Марія Василівна',
 'Коваленко Олександра Юріївна', 'Мельничук Альбіна Ігорівна ',
 'Шевченко Марія Григорівна', 'Вража Андрій Вікторович',
 'Ткач  Христина Олександрівна', 'Аврамов Михайло Іванович',
 'Курильчик Ксенія Максимівна'],
    'Дата народження': ['1985-05-15', '1990-08-20', '1978-03-10', '1982-11-25', '1995-07-30',
                       '1988-09-12', '1975-12-05', '1992-04-18', '1980-10-22', '1998-02-28'],
    'Посадовий оклад': [15000, 18000, 22000, 19000, 16000, 21000, 25000, 17000, 20000, 23000],
    'Стаж': [15, 8, 22, 12, 3, 10, 25, 6, 18, 2],
    'Відпрацьовано днів': [20, 22, 18, 21, 20, 23, 20, 19, 22, 20]
}

df = pd.DataFrame(data)
df['Дата народження'] = pd.to_datetime(df['Дата народження'])

# 2. Розрахунки
# Для прикладу візьмемо 21 робочий день у місяці
робочих_днів = 21

df['Нарахована ЗП'] = df['Посадовий оклад'] / робочих_днів * df['Відпрацьовано днів']
df['Доплата за стаж'] = df['Посадовий оклад'] * df['Стаж'] * 0.001
df['Всього нараховано'] = df['Нарахована ЗП'] + df['Доплата за стаж']
df['Податок'] = df['Всього нараховано'] * 0.195
df['До видачі'] = df['Всього нараховано'] - df['Податок']

# Збереження даних
df.to_csv('зарплата_співробітників.csv', index=False)

# 3. Статистичні розрахунки
середня_зп = df[df['Відпрацьовано днів'] == робочих_днів]['До видачі'].mean()
стаж_більше_10 = len(df[df['Стаж'] > 10])

print(f"\nСередня зарплата співробітників, які відпрацювали повний місяць: {середня_зп:.2f} грн")
print(f"Кількість співробітників зі стажем більше 10 років: {стаж_більше_10}")

# 4. Візуалізація
plt.figure(figsize=(12, 6))
plt.barh(df['ПІБ'], df['До видачі'], color='skyblue')
plt.title('Зарплата до видачі за поточний місяць', fontsize=14)
plt.xlabel('Сума, грн', fontsize=12)
plt.ylabel('Співробітник', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('зарплата.png', dpi=300)
plt.show()
```
